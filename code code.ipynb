```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data logger Januari 2024 /Unit Amikom/Unit 7/Unit 7 fix/20230228_PM5330.csv")
df.head()
```

```python
df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])
df['date'] = df['Time Stamp'].dt.date
```

```python
for tanggal in df['date'].unique():
    data_harian = df[df['date'] == tanggal]
    nama_file = f'data_logger_{tanggal}.csv'
    data_harian.to_csv(nama_file, index=False)
```
